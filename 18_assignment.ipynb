3) What is App routing in Flask? Why do we use app routes?

App routing in Flask is the process of mapping URLs to specific functions within a Flask application. It allows developers to handle incoming requests from clients and return appropriate responses. In other words, app routing helps Flask to identify the requested URL and route it to the appropriate view function to generate a response
1) URL Mapping : App routing is used to map URLs to specific functions within the Flask application. It allows developers to handle incoming requests from clients and return appropriate responses
2) Handling HTTP Requests :  Flask app routes help to handle different types of HTTP requests such as GET, POST, PUT, DELETE, etc. and perform the required actions accordingly
3) Organizing Code : App routes help developers to organize the code for their Flask application by keeping related functionality within the same view function

5) What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function 

In Flask, the url_for() function is used for URL building. This function generates a URL for the given endpoint with the specified method It takes the endpoint name as the first argument and any number of keyword arguments representing the variable parts of the URL

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the index page.'

@app.route('/user/<username>')
def profile(username):
    return f'This is the profile page for {username}.'

@app.route('/blog/<int:post_id>')
def blog(post_id):
    return f'This is the blog post {post_id}.'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index'))
        print(url_for('profile', username='Srinu'))
        print(url_for('blog', post_id=1, _external=True))
```
