2) Write a python program to print Python Exception Hierarchy

```python
def print_exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

print_exception_hierarchy(Exception)
```

The LookupError class is a built-in Python exception that serves as the base class for various lookup-related exceptions. It is raised when a key or index is not found during a lookup operation.

Some of the errors that are defined in the LookupError class include:

IndexError: Raised when an index is out of range for a sequence
KeyError: Raised when a key is not found in a dictionary.

6) List down some best practices for exception handling in python.

Catch specific exceptions: Catch only the exceptions that you expect to be raised, and let other exceptions propagate up the call stack. 
This can help you avoid catching and handling unexpected exceptions, which can lead to bugs and make it harder to debug your code.

Use try-except blocks sparingly: Use try-except blocks only where necessary, and keep them as short as possible. 
This can make your code easier to read and understand, and can also help you avoid catching and handling exceptions that you don't expect.

Provide helpful error messages: Provide error messages that are clear and concise, and that provide enough information for the user to understand 
what went wrong and how to fix it. This can help make your code more user-friendly, and can also make it easier to debug your code.

Use the finally clause: Use the finally clause to clean up resources, such as file handles or database connections, that were opened in a try block.
This can help ensure that resources are properly released, even if an exception is raised.

Don't ignore exceptions: Don't ignore exceptions, even if you don't know how to handle them. Ignoring exceptions can lead to bugs and make 
it harder to debug your code.

Don't catch Exception: Don't catch the Exception base class, as this can catch unexpected exceptions and make it harder to debug your code. 
Instead, catch only the specific exceptions that you expect to be raised.
