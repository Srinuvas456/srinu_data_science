7) You are a data analyst for a flight booking website, and you have been asked to analyze the Flight
Price dataset to identify any trends in flight prices. What features would you analyze to identify these
trends, and what visualizations would you use to present your findings to your team?

```python
df['Date_of_Journey']=pd.to_datetime(df['Date_of_Journey'])
monthly_prices = df.groupby(df['Date_of_Journey'].dt.month)['Price'].mean()
plt.plot(monthly_prices.index, monthly_prices.values)
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.title('Average Flight Prices by Month')
plt.show()
```

```python
route_prices = df.groupby(['Source', 'Destination'])['Price'].mean().reset_index()
sns.barplot(x='Price', y='Source', hue='Destination', data=route_prices)
plt.xlabel('Average Price')
plt.ylabel('Source')
plt.title('Average Flight Prices by Route')
plt.show()
```

```python
airline_prices = df.groupby('Airline')['Price'].mean().reset_index()
sns.barplot(x='Price', y='Airline', data=airline_prices)
plt.xlabel('Average Price')
plt.ylabel('Airline')
plt.title('Average Flight Prices by Airline')
plt.show()
```

```python
df['Dep_Time'] = pd.to_datetime(df['Dep_Time'])
df['Hour'] = df['Dep_Time'].dt.hour
time_prices = df.groupby('Hour')['Price'].mean()
plt.plot(time_prices.index, time_prices.values)
plt.xlabel('Hour')
plt.ylabel('Average Price')
plt.title('Average Flight Prices by Time of Day')
plt.show()
```

15) A company wants to launch a new app on the Google Playstore and has asked you to analyze the
Google Playstore dataset to identify the most popular app categories. How would you approach this
task, and what features would you analyze to make recommendations to the company?

```python
df.drop_duplicates(subset='App', keep='first', inplace=True)
df['Installs'] = df['Installs'].apply(lambda x: int(x.replace(',', '').replace('+', '')))
category_installs = df.groupby(['Category'])['Installs'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(category_installs.index, category_installs.values)
plt.xticks(rotation=90)
plt.xlabel('App Category')
plt.ylabel('Average Number of Installs')
plt.title('Most Popular App Categories on Google Playstore')
plt.show()
```
