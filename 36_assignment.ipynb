1) Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.

```python
import scipy.stats as stat
import numpy as np
sample_mean = 50
sample_std = 5
n = 100
std_error = sample_std /(n**0.5)
margin_error = stat.t.ppf(0.975,n-1) * std_error
lower = sample_mean - margin_error
upper = sample_mean + margin_error
print("The 95% confidence interval is ({:.2f}, {:.2f})".format(lower, upper))
```

2) Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.

```python
import numpy as np
from scipy.stats import chisquare

observed_freq = [48, 23, 17, 9, 8, 25] 
expected_freq = [60, 60, 60, 30, 30, 60] 

test_statistic, p_value = chisquare(observed_freq, expected_freq)


print("Chi-square test statistic: {:.2f}".format(test_statistic))
print("p-value: {:.4f}".format(p_value))
if p_value < 0.05:
    print("Reject the null hypothesis that the observed frequencies match the expected frequencies.")
else:
    print("Fail to reject the null hypothesis that the observed frequencies match the expected frequencies.")

```

4) A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
population who smoke.

```python
import math
from scipy.stats import norm
n = 500
p = 60 / n
z = norm.ppf(0.975) 
lower = p - z * math.sqrt(p * (1 - p) / n)
upper = p + z * math.sqrt(p * (1 - p) / n)
print("95% Confidence Interval: ({:.4f}, {:.4f})".format(lower, upper))
```

5) Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation
of 12 using Python. Interpret the results.

```python
import math
from scipy.stats import t
n = 100
xbar = 75
s = 12
t_score = t.ppf(0.95,n-1)
lower = xbar - t_score * s / math.sqrt(n)
upper = xbar + t_score * s / math.sqrt(n)
print("90% Confidence Interval: ({:.2f}, {:.2f})".format(lower, upper))
```

6) Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
df = 10
x = np.linspace(chi2.ppf(0.001, df), chi2.ppf(0.999,df), 1000)
flg, ax = plt.subplots()
ax.plot(x, chi2.ppf(x, df))
x_fill = np.linspace(chi2.ppf(0.15, df), chi2.ppf(0.999, df), 1000)
ax.fill_between(x_fill, chi2.pdf(x_fill, df), color='orange')
ax.set_xlabel('Chi-Square Statistic')
ax.set_ylabel('Probability Density')
ax.set_title('Chi-Square Distribution with 10 Degrees of Freedom')
plt.show()
```

7) A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who
prefer Coke.

```python
import math
from scipy.stats import norm
n = 1000
p_hat = 520/n
z_score = norm.ppf(0.99)
lower = p_hat - z_score * math.sqrt(p_hat * (1 - p_hat) / n)
upper = p_hat + z_score * math.sqrt(p_hat * (1 - p_hat) / n)
print("99% Confidence Interval: ({:.4f}, {:.4f})".format(lower, upper))
```

8) A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
expected frequencies of a fair coin. Use a significance level of 0.05.

To conduct a chi-square goodness of fit test to determine if the observed frequencies of a coin biased towards tails match the expected frequencies of a fair coin, given 100 flips and 45 tails, we can follow these steps
1) Define the null hypothesis and alternative hypothesis:

Null hypothesis (H0): The coin is fair, and the observed frequencies match the expected frequencies of a fair coin.
Alternative hypothesis (Ha): The coin is biased towards tails, and the observed frequencies do not match the expected frequencies of a fair coin.
2) Set the significance level (alpha) to 0.05.

3) Calculate the expected frequencies under the null hypothesis. For a fair coin, the expected frequency of tails is 0.5 and the expected frequency of heads is 0.5.

4) Calculate the chi-square test statistic using the formula:

X^2 = sum((O - E)^2 / E)

where:

O is the observed frequency of tails (45 in this case).
E is the expected frequency of tails (50 for a fair coin).
The sum is taken over all categories of the variable (tails and heads in this case).
5) Calculate the degrees of freedom (df) as the number of categories minus 1 (1 in this case).

6) Calculate the p-value using the chi-square distribution with df degrees of freedom and the calculated test statistic.
7) Compare the p-value to alpha. If the p-value is less than or equal to alpha, reject the null hypothesis and conclude that the coin is biased towards tails. If the p-value is greater than alpha, fail to reject the null hypothesis and conclude that there is insufficient evidence to conclude that the coin is biased towards tails.

```python
import scipy.stats as stat
alpha = 0.05
n = 100
p = 0.5
expected_tails = p * n
expected_heads = (1 - p) * n
observed_tails = 45
observed_heads = n - observed_tails
expected_freqs = [expected_tails, expected_heads]
observed_freqs = [observed_tails, observed_heads]
test_statistic, p_value = stat.chisquare(f_obs=observed_freqs, f_exp=expected_freqs)
df = len(expected_freqs) - 1
if p_value <= alpha:
    print("p-value = {:.4f}, Reject the null hypothesis. The coin is biased towards tails.".format(p_value))
else:
    print("p-value = {:.4f}, Fail to reject the null hypothesis. There is insufficient evidence to conclude that the coin is biased towards tails.".format(p_value))
```

9) A study was conducted to determine if there is an association between smoking status (smoker or
non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below.
Conduct a chi-square test for independence to determine if there is a significant association between
smoking status and lung cancer diagnosis.

Use a significance level of 0.05.
Group A

Outcome 1 20 15
Outcome 2 10 25
Outcome 3 15 20
Group B

Lung Cancer: Yes

Smoker 60 140
Non-smoker 30 170

```python
import numpy as np
import scipy.stats as stat
alpha = 0.05
table = np.array([[20, 15], [10, 25], [15, 20], [60, 140], [30, 170]])
row_totals = table.sum(axis=1)
col_totals = table.sum(axis=0)
grand_total = table.sum()
expected_freqs = np.outer(row_totals, col_totals) / grand_total
test_statistic, p_value, df, expected_freqs = stat.chi2_contingency(table)
if p_value <= alpha:
    print("p-value = {:.4f}, Reject the null hypothesis. There is a significant association between smoking status and lung cancer diagnosis.".format(p_value))
else:
    print("p-value = {:.4f}, Fail to reject the null hypothesis. There is insufficient evidence to conclude that there is a significant association between smoking status and lung cancer diagnosis.".format(p_value))
```

11) A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.

```python
from scipy.stats import ttest_1samp
sample_mean = 72
sample_std = 10
null_hypothesis = 70
alpha = 0.05

n = 30
t_statistic, p_value = ttest_1samp([sample_mean]*n, null_hypothesis)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
if p_value < alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```
