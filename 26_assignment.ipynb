1) What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a data visualization library for Python that allows you to create a wide range of static, animated, and interactive visualizations 
in Python. It is widely used for creating high-quality plots, charts, histograms, and other graphical representations of data
uses: 
1) Exploring data and understanding patterns.
2) Communicating insights to others in a clear and concise manner.
3) Analyzing trends and relationships between variables.
4) Comparing different data sets visually.
5) Creating publication-quality figures and images.
Matplotlib :
1) Lineplot : A simple plot of data points connected by straight lines, which is useful for visualizing trends and patterns over time
2) Scatterplot : A plot that shows the relationship between two variables by placing points on a two-dimensional plane, which is useful for identifying patterns and correlations between variables
3) Histogram : A plot that shows the distribution of a single variable, which is useful for identifying the range, skewness, and outliers of a variable
4) Bar plot : A plot that shows the distribution of a categorical variable, which is useful for comparing the frequency or proportion of different categories
5) Boxplot : A plot that shows the distribution of a variable using the quartiles and outliers, which is useful for identifying the median, range, and skewness of a variable

2) What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("This is a scatter plot")
plt.show()
```

3) Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])  
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
fig, axs = plt.subplots(2, 2, figsize = (10,8))
axs[0,0].plot(x1,y1)
axs[0,0].set_title('Line1')
axs[0,1].plot(x2,y2)
axs[0,1].set_title('Line2')
axs[1,0].plot(x3,y3)
axs[1,0].set_title('Line3')
axs[1,1].plot(x4,y4)
axs[1,1].set_title('Line4')
fig.suptitle('Four Line Plots')
plt.show()
```

4) What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.title("profit by company")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.xlabel("profit")
plt.ylabel("company")
plt.title("profit by company")
plt.show()
```

5) What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot is a graphical representation of a dataset that shows the distribution of the data by displaying the median, quartiles, and outliers. 
The box represents the middle 50% of the data, the line inside the box represents the median, and the whiskers represent the rest of the data
Box plots are used to compare the distribution of data across different groups or datasets, and to identify potential outliers or skewness in the data


```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)
data = [box1, box2]
plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```
