1) Explain GET and POST methods.

1) GET : This method is used to request data from a server. When you enter a URL in your browser and press enter, a GET request is 
sent to the server asking for the specified web page. The server responds with the requested page, which is then displayed in your 
browser. The GET method can also be used to retrieve data in a variety of formats, such as JSON, XML, or HTML
2) POST : This method is used to send data to a server. When you fill out a form on a website and submit it, a POST request is 
sent to the server containing the data you entered. The server then processes the data and may send a response back to your browser. 
POST requests are commonly used to create or update data on a server, such as submitting a blog post or updating a user profile

2) Why is request used in Flask?

1) Request method: whether it is a GET, POST, PUT, DELETE or any other HTTP method.
2) Request headers: such as the User-Agent, Accept, and Content-Type headers.
3) Request data: any data that was submitted with the request, such as form data or JSON payload.
4) Request parameters: any query parameters that were included in the URL

3) Why is redirect() used in Flask?

redirect(): It is a function used to redirect a user to a different URL. It is a built-in function in the flask module and is 
commonly used to redirect users to a different page after a form submission, after a successful login, or after completing any 
other action that requires a redirect

4) What are templates in Flask? Why is the render_template() function used?

IN flask emplates are used to create dynamic HTML pages that can be customized based on user input or data retrieved from a database. 
Templates are typically created using a templating engine such as Jinja2, which allows for variables, control structures, and inheritance
Templates in Flask are typically stored in a separate directory called "templates" and can be rendered using the render_template() function,
which is a built-in function in Flask. The render_template() function takes the name of the template as its argument and returns the HTML
content of the rendered template
